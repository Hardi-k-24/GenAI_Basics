```python
from google import genai

client = genai.Client()

response=client.models.generate_content(
    model="gemini-2.5-flash",
    contents='Explain how AI works in few words'
)
print(response.text)
```

**Using Google Gemini API for chat completion (Multi Turn Conversation)**

```python
from google import genai

client=genai.Client()

chat = client.chats.create(model="gemini-2.5-flash")

response = chat.send_message("I have 2 dogs in my house.")
print(response.text)

print("-"*100)

response = chat.send_message("How many paws are in my house?")
print(response.text)
```

**Using GROQ API and Chat Completion(more like openai format)**

```python
!pip install groq
```

```python
import os
```

```python
os.environ["GROQ_API_KEY"]=""
```

```python
from groq import Groq

client=Groq()

chat_completion=client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role":"system","content":"you are a helpful assistant"},
        {"role":"user","content":"add 2+2"},
        {"role":"assistant","content":""},
        {"role":"user","content":"add 2 more to it"}
    ]
  )
print(chat_completion.choices[0].message.content)
```
